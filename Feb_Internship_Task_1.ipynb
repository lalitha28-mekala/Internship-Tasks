2. Pass / Fail Analyzer
Problem Statement: Analyze student results. marks = [45, 78, 90, 33, 60]

- Requirements:
- A student passes if marks â‰¥ 50
- Count the total number of pass students
- Count the total number of fail students
- Print the final result clearly


```python
Marks=[45,78,90,33,60]
pass_students=0
fail_students=0
for i in Marks:
  if i>=50:
    pass_students +=1
  else:
    fail_students +=1
print("Total Number of Pass Students :",pass_students)
print("Total Number of fail Students :",fail_students)
```

3. Simple Data Cleaner ðŸ‡±Problem Statement:Clean and standardize user names.
- names = [" Alice ", "bob", " CHARLIE "]

- Requirements:
- Remove extra spaces
- Convert all names to lowercase
- Print the cleaned list
- Real-World Application: Data preprocessing before analysis


```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names=[]
for i in names:                # strip() removes all spaces at start and ending of words
  i=i.strip().lower()
  cleaned_names.append(i)
print(cleaned_names)
```

4. Message Length Analyzer Problem Statement: Analyze message sizes.
- messages = ["Hi", "Welcome to the platform", "OK"]

- Requirements:
- Print the length of each message
- Flag messages longer than 10 characters
- Real-World Application: Text filtering and validation systems

```python
messages = ["Hi", "Welcome to the platform", "OK"]
print("Message  : Length")
for i in messages:
  print(f"{i} :{len(i)}")
  if len(i)>10:
    print("This Message is Longer than 10 Characters!")
  else:
    print("This is Short Message")
```

Q 5 Problem Statement: Detect error messages from system logs.
- logs = ["INFO", "ERROR", "WARNING", "ERROR"]

- Requirements:
- Count the number of "ERROR" entries
- Print the total error count


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
  if i=="ERROR":
    count +=1
print("Error was appearied in :",count)
```
