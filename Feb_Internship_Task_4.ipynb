## Problem 1: Mobile Recharge Validation System
- Problem Statement : Create a function to validate a mobile recharge request.
- Rules:Recharge amount must be ≥ ₹50
 - Valid plans: 199, 299, 399, 599
- User should be allowed to retry if invalid
- Real-Time Use :Telecom recharge apps,Payment gateways
- Hint :
  - Use function for validation
  - Use while loop for retry


```python
def mobile_recharge_validation(amt):
  valid_recharge_plans=[199,299,399,599]
  if amt >= 50 and amt in valid_recharge_plans:
    return True
  else:
    return  False

while True:
  amt=int(input("Enter Recharge Amount :"))
  if mobile_recharge_validation(amt):   # If function is True it prints successful and break the loop
    print("Recharge Successful")
    break
  else:
    print("Please Enter Valid Recharge Amount!")
```

## Problem 2: Inventory Reorder Alert System
- Problem Statement:Given a dictionary of products and stock quantities:
- Print Reorder Alert if stock < 15
- Print Stock OK otherwise
- Real-Time Use
  - Warehouse systems
  - Retail management
- Hint
  - Use function
  - Use for loop on dictionary


```python

# Function
def Inventory_Management(stock):
    for i,j in stock.items():
      if j <15 :
        print(f"{i}:{j}, Reorder Alert")
      else:
        print(f"{i}:{j},Stock ok")

inventory = {
    "Rice": 25,
    "Wheat": 10,
    "Sugar": 18,
    "Oil": 8,
    "Salt": 30}

# Call function
Inventory_Management(inventory)
```

## Problem 3: Student Result Processing System
- Problem Statement :Create a function that Accepts a list of student marks
- Calculates average marks Prints Pass if average ≥ 50 else Fail
- Real-Time Use Online exam systems,Evaluation portals
- Hint
    - Use loop to calculate total
    - Return result from function






```python
def results_evaluation(marks_list):
  total=0
  for i in marks_list:
    total +=i

  avg=total/(len(marks_list))
  if avg >= 50:
    return "Pass"
  else:
    return "Fail"
marks=[25,78,85,95,62,78]
result=results_evaluation(marks)
print(result)
```

## Problem 4: Cab Fare Estimator with Retry Option
- Problem Statement: Create a cab fare calculator:
- Base fare = ₹50,₹12 per km,Peak hour → 25% extra
Allow retry using while loop
- Real-Time Use
Ride-hailing apps,
Transport systems
- Hint
Function for fare calculation,
While loop for retry


```python
def cab_fare(kms,time):
  base_fare=50
  total_price= base_fare+(kms*12)
  if time=="peak":
    total_price = total_price+ total_price*0.25
    return total_price
  else:
    total_price = total_price
    return total_price

kms=int(input("Enter kilometers :"))
time=input("Select time peak/normal : ").lower()

result=cab_fare(kms,time)
print("Final fare:",result)
```

## Problem 5: Employee Attendance Eligibility Checker
- Problem Statement : Create a function that:Accepts attendance list ("P", "A")
- Calculates attendance percentage
- Returns Eligible if ≥ 75% else Not Eligible
- Real-Time Use: HR payroll systems,Internship tracking
- Hint:Use loop to count present days,Use return statement


```python
def attendance_eligibility_check(attendance):
  present_days=0
  for i in attendance:
    if i=="P":
      present_days +=1
  percentage=(present_days*100)/len(attendance)
  if percentage >=75:
    return f"Eligible,Percentage :{percentage:.2f}"
  else:
    return f"Not Eligible,Percentage :{percentage:.2f}"
# Call function
result=attendance_eligibility_check(["P","A","P","A","P","A","P","A"])
print(result)
```

## Problem 6: Password Strength Checker
- Problem Statement : Create a function to check password strength.
- Rules:Minimum length = 8, Must contain at least one digit,Must contain at least one special character (@#$)
- Real-Time Use:Login systems,Security validation
- Hint: Use loop over string Use conditions inside function


```python
def password_checker(password):
  has_digit=False
  has_special=False
  special_chrs=["@","$","#"]
  if len(password) <8:
    return "weak password minimum length should be 8 characters!"
  for i in password:
    if i.isdigit():
       has_digit = True
    if i in special_chrs:
       has_special = True
  if has_digit and has_special:
    return "Strong Password"
  else:
    return "Weak password try agin with one didgit and one special characters!"

print(password_checker("lalitha1@456"))
```
